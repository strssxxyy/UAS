```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv ('stock.csv')
```

```python
top_dividend = (df.groupby('Stock Name')['Dividend/Share']
                 .sum()
                 .reset_index()
                 .sort_values(by='Dividend/Share', ascending=False)
                 .head(5))
print(top_dividend)
```

```python
top_marketvalue = (df.groupby('Stock Name')['Market Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Market Value', ascending=False)
                 .head(5))
print(top_marketvalue)
```

```python
top_annualyield = (df.groupby('Stock Name')['Annual Yield']
                 .sum()
                 .reset_index()
                 .sort_values(by='Annual Yield', ascending=False)
                 .head(7))
print(top_annualyield)
```
